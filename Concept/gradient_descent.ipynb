```python
#import libraries

import numpy as np
import matplotlib.pyplot as plt
from numpy import random as r
%matplotlib inline
```

```python
#prepare X and Y using random

X = r.rand(100)
Y = 0.5*X + 0.5 #we can compare with W, b that we find, using gradient descent
```

```python
#plotting original data
plt.figure(figsize=(10,8))
plt.scatter(X, Y)
plt.title('x and y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def plot_pred(pred, y, num):
  plt.figure(figsize=(6,4))
  graphnum = int(num / 10) + 1
  plt.title('Graph {}'.format(graphnum))
  plt.scatter(X,y) #original graph
  plt.scatter(X,pred) #predicted graph using gradient descent
```

```python
#implement gradient descent

w = r.uniform(-1, 1)
b = r.uniform(-1, 1)

lr = 0.1 #learning rate

for epoch in range(1000):
  Y_pred = w * X + b

  error = np.abs(Y_pred - Y).mean()

  if error < 0.001:
    break   #loop can be stopped if error is small enough
  
  w_grad = lr * ((Y_pred - Y)*X).mean()
  b_grad = lr * (Y_pred - Y).mean()

  #update w and b
  w = w - w_grad
  b = b - b_grad

  if epoch % 10 == 0:
    plot_pred(Y_pred, Y, epoch) #plot graph every 10 times
    
```

```python
print('Parameters w:: {}, b: {}'.format(w, b))
```
